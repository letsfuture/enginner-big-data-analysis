# 04. Apply, Map

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv")
```

```python
df.head()
```

```python
# 데이터의 행과 열의 갯수
df.shape
```

```python
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'   
}
```

```python
df['newIncome'] = df['Income_Category'].map(lambda x: dic[x])
df
```

```python
# Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑
def changeCategory(x):
    if x=='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'
```

```python
df['newIncome'] = df['Income_Category'].apply(changeCategory)
df
```

```python
# Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의
df['AgeState'] = df['Customer_Age'].map(lambda x: x//10*10)
df['AgeState'].value_counts().sort_index()
```

```python
# Education_Level의 값 중 Graduate단어가 포함되는 값은 1, 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력
df['newEduLevel'] = df['Education_Level'].map(lambda x: 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts().sort_index()
```

```python
# Credit_Limit 컬럼값이 4500 이상인 경우 1, 그외의 경우에는 모두 0으로 하는 newLimit 
df['newLimit'] = df['Credit_Limit'].map(lambda x: 1 if x>=4500 else 0)
df['newLimit'].value_counts().sort_index()
```

```python
# Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum 인 경우 1, 그외의 경우에는 모두 0으로 하는 newState
def check(x):
    if x['Marital_Status']=='Married' and x['Card_Category'] == 'Platinum':
        return 1
    else:
        return 0
        
df['newState'] = df.apply(check, axis=1)
df['newState'].value_counts().sort_index()
```

```python
# Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의하고
df['Gender'] = df['Gender'].map(lambda x: 'male' if x=='M' else 'female')
df['Gender'].value_counts().sort_index()
```

# 05. Time Series

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/timeTest.csv")
```

```python
df.head()
```

```python
# Yr_Mo_Dy을 판다스에서 인식할 수 있는 datetime64타입으로 변경
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
df.info()
```

```python
# Yr_Mo_Dy에 년도가 2061년 이상의 경우 100을 빼서 새롭게 날짜를 Yr_Mo_Dy 컬럼에 정의
def fix_century(x):
    import datetime
    
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))
```

```python
df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fix_century)
df.head()
```

```python
# 년도별 각 컬럼의 평균값
# df.groupby(df['Yr_Mo_Dy'].dt.year).mean(numeric_only=False)
df.groupby(df['Yr_Mo_Dy'].dt.year).mean(numeric_only=True)
```

```python
# weekday컬럼을 만들고 요일별로 매핑
df['weekday'] = df['Yr_Mo_Dy'].dt.weekday
df
```

```python
# weekday컬럼을 기준으로 주말이면 1, 평일이면 0의 값을 가지는 WeekCheck 컬럼
df['weekCheck'] = df['weekday'].map(lambda x: 1 if x in [5, 6] else 0)
df['weekCheck'].to_frame()
```

```python
def isWeekend(x):
    if x == 5 or x == 6:
        return 1
    else:
        return 0
    
df['weekCheck'] = df['weekday'].apply(isWeekend)
df['weekCheck'].to_frame()
```

```python
# 년도, 일자 상관없이 모든 컬럼의 각 달의 평균
df.groupby(df['Yr_Mo_Dy'].dt.month).mean(numeric_only=True)
```

```python
# 모든 결측치는 컬럼 기준 직전의 값으로 대체하고, 첫 번째 행에 결측치가 있을 경우 뒤에 있는 값으로 대체
print('변경 전', df.isnull().sum(), sep='\n')
print('\n')

df = df.fillna(method='ffill').fillna(method='bfill')

print('변경 후', df.isnull().sum(), sep='\n')
```

```python
# 년도 - 월을 기준으로 모든 컬럼의 평균값
df.groupby(df['Yr_Mo_Dy'].dt.to_period('M')).mean(numeric_only=True)
```

```python
# RPT 컬럼의 값을 일자별 기준으로 1차 차분
df['RPT'].diff()
```

```python
# RPT와 VAL의 컬럼을 일주일 간격으로 각각 이동평균한 값
df[['RPT', 'VAL']].rolling(7).mean()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv")
```

```python
# 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경
def changeDate(x):
    import datetime
    
    hour = x.split(':')[1]
    date = x.split(':')[0]
    
    if hour == '24':
        hour = '00:00:00'
        finalDate = pd.to_datetime(date + ' ' + hour) + datetime.timedelta(days=1)
    else:
        hour = hour + ':00:00'
        finalDate = pd.to_datetime(date + ' ' + hour)
    return finalDate
```

```python
df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(changeDate)
df
```

```python
# 일자별 영어요일 이름을 dayName 컬럼에 저장
df['dayName'] = df['(년-월-일:시)'].dt.day_name()
df
```

# 07. Merge, Concat

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv', index_col= 0)
```

```python
df.head()
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

```python
# df1과 df2 데이터를 하나의 데이터프레임으로
pd.concat([df1, df2])
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
# df3과 df4 데이터를 하나의 데이터프레임으로
# 단, 둘 다 포함하고 있는 년도에 대해서만 고려
pd.concat([df3, df3], join='inner')
```

```python
# df3과 df4 데이터를 하나의 데이터프레임으로
# 단, 모든 컬럼을 포함하고 결측치는 0으로 대체
pd.concat([df3, df4], join='outer').fillna(0)
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

```python
# df5과 df6 데이터를 하나의 데이터프레임으로(merge함수 이용)
# 단, Algeria 컬럼을 key로 하고, 두 데이터 모두 포함되는 데이터만 출력
pd.merge(df5, df6, on='Algeria', how='inner')
```

```python
# Algeria 컬럼을 key로 하고, 합집합인 데이터만 출력
pd.merge(df5, df6, on='Algeria', how='outer')
```

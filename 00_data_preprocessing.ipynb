```python
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv")
```

**각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력:** `groupby().size()`, `value_counts().sort_index()`

```python
Ans = df.groupby('host_name').size()
Ans
```

```python
Ans = df['host_name'].value_counts().sort_index()
Ans
```

**각 host_name의 빈도수를 구하고, 빈도수 기준 내림차순 정렬한 데이터프레임 생성:** `size()`, `to_frame()`, `sort_values()`

```python
Ans = df.groupby('host_name').size().to_frame()
Ans 
```

```python
Ans.rename(columns={0: 'counts'}).sort_values('counts', ascending=False)
```

**neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수:** `groupby()`, `size()`

groupby를 사용하면 기본으로 그룹 라벨이 index가 된다.<br>
index를 사용하고 싶지 않은 경우에는 `as_index=False`를 설정하면 된다.

```python
Ans = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
Ans
```

**neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력**

```python
df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().groupby(['neighbourhood_group'], as_index=False).max()
```

**참고**<br>
`get_group()`: 그룹 안에 데이터를 확인하고 싶은 경우 사용

```python
df.groupby('neighbourhood').get_group('Woodlawn')
```

**neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값:** `groupby().agg()`

```python
df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```

**neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값:** `groupby().agg()`

```python
df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```

**neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균:** `df.groupby(['column1', 'column2'])['컬럼3'].agg()`

```python
df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].agg(['mean'])
```

**neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하라:** `unstack()`

```python
df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].mean().unstack()
```

**neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 채워라**

```python
df.groupby(['neighbourhood', 'neighbourhood_group'])['price'].mean().unstack().fillna(-999)
```

**neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라**

```python
df[df['neighbourhood_group'].str.contains('Queens')].groupby(['neighbourhood'])['price'].agg(['mean', 'var', 'max', 'min'])
```

```python
# 정답
df[df['neighbourhood_group']=='Queens'].groupby(['neighbourhood'])['price'].agg(['mean', 'var', 'max', 'min'])
```

**neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율**

```python
df[['neighbourhood_group', 'room_type']].groupby(['neighbourhood_group', 'room_type']).size()
```

```python
df[['neighbourhood_group', 'room_type']].groupby(['neighbourhood_group', 'room_type']).size().unstack()
```

```python
Ans = df[['neighbourhood_group', 'room_type']].groupby(['neighbourhood_group', 'room_type']).size().unstack()
Ans.loc[:, :] = (Ans.values / Ans.sum(axis=1).values.reshape(-1, 1))
Ans
```
